```python
import pandas as pd

# 엑셀 파일 경로 설정
file_path = 'userdata.xlsx'
df = pd.read_excel(file_path)

name = df['1. 이름'].values
depart = df['2. 학과'].values
id = df['3. 학번'].values


pwd = []
for i in range(len(id)):
    tmp_id = str(id[i])
    dep = depart[i]
    print(dep)
    if dep == '인공지능응용학과':
        pw = 'aai'
    elif dep == '지능형반도체공학과':
        pw = 'dse'
    elif dep == '미래에너지융합학과':
        pw = 'eng'
    tmp = 'ccc'+pw+tmp_id[4:]
    print(tmp)
    pwd.append(tmp)
    
```
